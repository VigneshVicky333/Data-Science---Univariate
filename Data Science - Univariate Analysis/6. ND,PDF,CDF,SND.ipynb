# Next we need to find the "Probability Density function".

# This (PDF) is not an In-Built function, so we need to write the function for it.


```python
# We have already written the code for this, we can use it (I searched and found it from our G-Drive link)

# https://drive.google.com/drive/folders/1XMrQpGmmmOQQYeeu6rbwZOAhNdndCysG?ths=true
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
# If we need to write a function for finding the values from 60 to 70. It automatically calculates the probaliltiy density function.
```

```python
# get_pdf_probability(dataset,startrange,endrange):

get_pdf_probability(dataset["ssc_p"],60,70)
```

# Standard Normal Distribution

```python
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```

### Explanation of the Code:

#### 1. **Importing Library**:
- `import seaborn as sns`: Import the Seaborn library for visualization.

#### 2. **Calculate Mean and Standard Deviation**:
- `mean = dataset.mean()`: Calculates the average value of the dataset.
- `std = dataset.std()`: Calculates the spread of the dataset.

#### 3. **Standardize the Dataset**:
- `values = [i for i in dataset]`: Creates a list of all values in the dataset.
- `z_score = [((j - mean) / std) for j in values]`: 
  - This is a **one-liner for loop** that calculates the Z-score for each value in the dataset.
  - **Z-score formula**: 
    \[
    Z = \frac{X - \mu}{\sigma}
    \]
    Where:
    - \( X \): Value in the dataset.
    - \( \mu \): Mean of the dataset.
    - \( \sigma \): Standard deviation of the dataset.

#### 4. **Plot the Standardized Data**:
- `sns.distplot(z_score, kde=True)`:
  - Generates a distribution plot for the Z-scores using Seaborn.
  - The parameter `kde=True` ensures the graph includes a Kernel Density Estimate (smooth curve) along with the histogram.

#### 5. **Return Mean of Z-Scores**:
- `sum(z_score) / len(z_score)`:
  - Calculates the mean of the Z-scores.
  - For a standard normal distribution, this value should be close to 0.

### Summary:
This code:
1. Calculates the Z-scores for a dataset to standardize it.
2. Visualizes the standardized data using Seaborn's `distplot`.
3. Verifies the standardization by calculating the mean of the Z-scores, which should ideally be 0.


```python
stdNBgraph(dataset["ssc_p"])
```
